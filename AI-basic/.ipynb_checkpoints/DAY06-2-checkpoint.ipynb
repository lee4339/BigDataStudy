```python
'''
UCI REPO - Adult Data Set
adult.data
나이가 짝수인 남성은 몇명인가
'''

with open('adult.data', 'r') as f:
    raw = f.readlines()
print(raw[:5])
```

```python
'''
raw[0] --> 문자열 한줄, 첫번째 사람에 대한 정보
raw[0].split(',') --> 리스트(문자열을 ',' 기준으로 분리, 리스트화)
enumerate(raw[0].split(',')) --> 리스트를 순환, 인덱스, 리스트 인덱스별 값을 tuple로 반환
'''

first_person_list = raw[0].split(',')
for index, person_info in enumerate(first_person_list):
    print(index, person_info) #나이 -- 0, 성별 -- 9
```

```python
filtered_data = []
for person_string in raw:
    person_list = person_string.split(',')
    try:
        age = person_list[0]
        sex = person_list[9]
        filtered_data.append([age, sex])
    except:
        pass
    
print(filtered_data)
```

```python
for p in filtered_data:
    p[0] = int(p[0])
    p[1] = p[1].strip()
print(filtered_data)
```

```python
count = 0
for p in filtered_data:
    if p[0]%2 == 0 and p[1] == 'Male':
        count = count + 1
print(count)
```
