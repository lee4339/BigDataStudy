```python
def fn1(n):
    sum = 0
    for i in range(1, n + 1):
        sum = sum + (2 * i + 1)
    return sum

print(fn1(100)) # i : 1 ~ 100
```

```python
#빅 파이
#n까지의 수를 전부 곱해준다.

def fn2():
    r = 1
    for i in range(1, 3):
        for j in range(4, 7):
            r = r * (3 * i * j)
    return r

fn2()
```

```python
#정규분포 수식

def normal_dist(x, mu = 0, sigma = 1):
    v1 = ((x - mu) / sigma)**2
    v2 = math.exp(-0.5 * v1)
    v3 = 1/(sigma * math.sqrt(2 * math.pi))
    return v2 * v3
```

```python
def test_fn(x):
    return x * 2 + 5

import matplotlib.pyplot as plt

list_x = [x/100.0 for x in range(-500, 500)]
list_y = [normal_dist(x) for x in list_x]
list_y2 = [normal_dist(x, 0, 0.5) for x in list_x]
#print(list_x)
#print(list_y)

_, axe = plt.subplots()  #튜플형태인데 _값에는 한 그래프 안에 몇등분할건지(ex)한면을 4등분하여 여러 그래프를 넣을 수 있음 그때는 pig라는 변수를 넣어주나 지금은 필요없으므로 _ 빈값)
axe.plot(list_x, list_y)
axe.plot(list_x, list_y2)
```
