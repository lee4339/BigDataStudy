```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('penguins')

df.head()
```

```python
df['species'].unique()
df['species'].value_counts()
```

```python
sns.displot(data=df, x='species')
sns.displot(data=df, x='island')
```

```python
fig = plt.figure(figsize=[12,4], facecolor='y')

axe_left = fig.add_subplot(1,2,1)
axe_left.set_title('Species')
sns.histplot(data=df, x='species', ax=axe_left, facecolor='red')

axe_right = fig.add_subplot(1,2,2)
axe_right.set_title('Island')
sns.histplot(data=df, x='island', ax=axe_right)
```

```python
fig2 = plt.figure(figsize=[20,4], facecolor='lavender')
fig2.suptitle('Bill_length_mm')

axe_left = fig2.add_subplot(1,3,1)
axe_left.set_title('By Species')
sns.histplot(data=df, x='bill_length_mm', hue='species', ax=axe_left, palette=['red','green','blue'])

axe_mid = fig2.add_subplot(1,3,2)
axe_mid.set_title('By Island')
sns.histplot(data=df, x='bill_length_mm', hue='island', kde=True, multiple='dodge', ax=axe_mid)

axe_right = fig2.add_subplot(1,3,3)
axe_right.set_title('By Sex')
sns.ecdfplot(data=df, x='bill_length_mm', hue='sex', ax=axe_right)
```
