```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('tips')
# df = sns.load_dataset('penguins')
# df = sns.load_dataset('diamonds')

print(df.head())
```

```python
# sns.displot(data=df, x='total_bill', binwidth=1)  # binwidth: bin(바의크기)의 폭
sns.displot(data=df, x='total_bill', bins=5)   # bins: bar의 갯수
a = pd.cut(df['total_bill'], bins=5)
print(a)
a.unique()
b = a.value_counts().sort_index()
_, axe = plt.subplots()
axe.bar(x=np.arange(len(b.index)), height=b.to_numpy(), width=1.0, edgecolor='black')
```

```python
# sns.displot(data=df, x='day')
# sns.displot(data=df, x='size', discrete=True)
sns.displot(data=df, x='day', shrink=0.7)
```

```python
# sns.displot(data=df, x='total_bill', hue='day')
# sns.displot(data=df, x='total_bill', hue='smoker')
# sns.displot(data=df, x='total_bill', hue='smoker', element='step') # bars, step, poly
# sns.displot(data=df, x='total_bill', hue='smoker', multiple='fill')  # layer, dodge, stack, fill
sns.displot(data=df, x='total_bill', col='smoker', hue='sex', multiple='dodge')
```

```python
# sns.displot(data=df, x='total_bill', stat='probability')
sns.displot(data=df, x='total_bill', hue='smoker', stat='probability', common_norm=False, multiple='dodge')
```

```python
# sns.displot(data=df, x='total_bill', kind='kde', bw_adjust=1, hue='sex', multiple='stack')
sns.displot(data=df, x='total_bill', kind='kde', bw_adjust=1, hue='sex', fill=True)
# sns.displot(data=df, x='total_bill', kind='kde', bw_adjust=1, hue='sex', cut=0)
# sns.displot(data=df, x='total_bill', kde=True)
```

```python
sns.displot(data=df, x='total_bill', kind='ecdf', hue='sex')
```

```python
# sns.displot(data=df, x='total_bill', y='tip')
# sns.relplot(data=df, x='total_bill', y='tip')
# sns.displot(data=df, x='total_bill', y='tip', binwidth=(2, 0.5), cbar=True)
sns.displot(data=df, x='total_bill', y='tip', kind='kde', thresh=0.1, levels=5, rug=True)
```

```python
# sns.jointplot(data=df, x='total_bill', y='tip')
# sns.jointplot(data=df, x='total_bill', y='tip', kind='reg')
j = sns.jointplot(data=df, x='total_bill', y='tip', kind='scatter')
j.plot_joint(sns.scatterplot)
j.plot_marginals(sns.boxplot)
```

```python
sns.rugplot(data=df, x='total_bill', y='tip')   #  혼자 쓰기 어려움
```

```python
p = sns.pairplot(data=df)
p.map_upper(sns.histplot)
p.map_lower(sns.kdeplot)
p.map_diag(sns.histplot, kde=True)
```
