```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('penguins')
print(df.head())

# island vs body_mass, hue=sex : strip. swarm
# species vs body_mass, hue=island: box, violin, boxen
```

```python
fig = plt.figure(figsize=[12,4])
a = fig.add_subplot(1,2,1)
sns.stripplot(data=df, x='island', y='body_mass_g', hue='sex', ax=a)

b = fig.add_subplot(1,2,2)
sns.swarmplot(data=df, x='island', y='body_mass_g', hue='sex', ax=b)
```

```python
fig = plt.figure(figsize=[20,4])
a = fig.add_subplot(1,3,1)
sns.boxplot(data=df, x='species', y='body_mass_g', hue='island', ax=a)

b = fig.add_subplot(1,3,2)
sns.violinplot(data=df, x='species', y='body_mass_g', hue='island', ax=b)

c = fig.add_subplot(1,3,3)
sns.boxenplot(data=df, x='species', y='body_mass_g', hue='island', ax=c)
```

```python
# heatmap: species, island, body_mass_g (mean)

a = df[['species', 'island', 'body_mass_g']].dropna()
print(a)

b = a.groupby(['species', 'island']).mean()
b = b.reset_index()
print(b)

c = b.pivot('species', 'island', 'body_mass_g')
sns.heatmap(c, annot=True, fmt='g')
```
