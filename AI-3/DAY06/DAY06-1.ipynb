```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd_data = pd.read_csv('전국교통약자이동지원센터정보표준데이터.csv', encoding='cp949')
print(pd_data.columns)

for i, n in enumerate(pd_data.columns):
    print(i, n)
```

```python
data = pd.DataFrame()

################################

s_addr1 = pd_data['소재지도로명주소']
s_addr2 = pd_data['소재지지번주소']
# print(s_addr1.hasnans)

# print(s_addr2[s_addr2.isna()])
s_addr2[s_addr2.isna()] = s_addr1[s_addr2.isna()]

s_addr3 = s_addr2.astype('str').map(lambda x: x.split(' ')[0])
# print(s_addr3.unique())
s_addr3.replace('전남', '전라남도', inplace=True)
s_addr3.replace('진안군', '전라북도', inplace=True)
data.insert(0, 'addr', s_addr3)

# s_addr1.astype('str').map(lambda x: x.split(' ')[0])
# print(s_addr1)

# s_addr2 = pd_data['소재지지번주소'].astype('str').map(lambda x: x.split(' ')[0])
# print(s_addr2)
# s_addr2.replace('nan', np.NaN, inplace=True)
# print(s_addr2.hasnans)
# s_addr2[s_addr2.isna()]

##################################

print(data)
```

```python
s_vcount = pd_data['보유차량대수']
# print(s_vcount.unique())
# print(s_vcount.dtype)
# print(s_vcount.hasnans)

data.insert(1, 'vcount', s_vcount)

data
```

```python
s_phone = pd_data['예약접수전화번호']
# print(s_phone.unique())
# print(s_phone.dtype)
# print(s_phone.hasnans)
data.insert(2, 'phone', s_phone)
print(data)
```

```python
def myf(x):
    if type(x) is type('str'):
        if x == 'x' or '없음' in x or '불가' in x or '안됨' in x:
            return np.NaN
        else:
            return x
    else:
        return x
    
def myf2(x):
    if type(x) is type('str') and len(x) < 10:
        return np.NaN
    else:
        return x

s_web = pd_data['예약접수인터넷주소']
s = s_web.map(myf2)
print(s.unique())

data.insert(3, 'web', s)
```

```python
def myf3(x):
    if type(x) is type('str'):
        if len(x)==1 or '없음' in x or '불가' in x or '안됨' in x or '미정' in x or  "00" in x :
            return np.NaN
        else:
            return x
    else:
        return x

s_app = pd_data['앱서비스명']
# print(s_app.unique())

s = s_app.map(myf3)
print(s.unique())
data.insert(4, 'app', s)

print(data)
```
