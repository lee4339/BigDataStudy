```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
```

```python
X, y = load_diabetes(return_X_y=True)
X = X[:,2][:, np.newaxis]   # bmi, 2d ndarry

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
m = LinearRegression().fit(X_train, y_train)
print(m.score(X_test, y_test))
print(m.coef_, m.intercept_)

_, axe = plt.subplots()
axe.scatter(X, y)

xs = np.linspace(-0.1, 0.15, 10)
ys = (m.coef_ * xs) + m.intercept_
axe.plot(xs, ys, c='r')
```

```python
#####
model = keras.Sequential([
    keras.layers.InputLayer(input_shape=(1,)),
    keras.layers.Dense(units=1)
])
model.summary()
print('weights', model.weights)
print('input', X_train[:10])
print('output', model.predict(X_train[:10]))

#model.compile(loss='mse', optimizer='SGD')
model.compile(loss='mse', optimizer=tf.optimizers.Adam(learning_rate=0.1))
history = model.fit(X_train, y_train, epochs=1000)

_, axe = plt.subplots()
axe.plot(history.history['loss'])

_, axe = plt.subplots()
axe.scatter(X, y)
xs = np.linspace(-0.1, 0.15, 10)
ys = model.predict(xs)
axe.plot(xs, ys, c='r')
```

```python
model2 = keras.Sequential([
    keras.layers.InputLayer(input_shape=(1,)),
    keras.layers.Dense(units=64, activation='relu'),
    keras.layers.Dense(units=64, activation='relu'),
    keras.layers.Dense(units=1)
])

model2.summary()
#model2.compile(loss='mse', optimizer=tf.optimizers.Adam(learning_rate=0.1))
#model2.compile(loss='mse', optimizer='Adadelta')
model2.compile(loss='mse', optimizer=tf.optimizers.Adadelta(learning_rate=0.1))
model2.fit(X_train, y_train, epochs=10)

_, axe = plt.subplots()
axe.scatter(X, y)
xs = np.linspace(-0.1, 0.15, 10)
ys = model2.predict(xs)
axe.plot(xs, ys, c='r')
```
