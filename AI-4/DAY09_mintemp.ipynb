```python
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import layers


pd_data = pd.read_csv('종관기상_지점별_연·월_통계_20221130120557.csv', encoding='cp949')
# print(pd_data.head())

wt = pd_data.T
wt.drop(0, axis=1, inplace=True)
wt.drop('관측지점별', axis=0, inplace=True)
# print(wt)


wt_np = wt.iloc[:,0].values.astype('float32')
# print(wt_np)
wt_np_x = wt_np.reshape(22, 12, 1)
```

```python
_, axe = plt.subplots()
axe.plot(wt_np)
```

```python
_, axe = plt.subplots()
# axe.plot(wt_np)
axe.plot(wt_np_x[1])
```

```python
X_train, y_train, X_test, y_test = wt_np_x[:21, :11], wt_np_x[:21, 11:], wt_np_x[21:, :11], wt_np_x[21:, 11:]
```

```python
model = keras.Sequential([
    layers.SimpleRNN(11, input_shape=(None, 1)),
    layers.Dense(1)
])

model.compile(loss='mse', optimizer=keras.optimizers.Adam(learning_rate=0.01))
hist = model.fit(X_train, y_train, epochs=100, verbose=0)
model.predict(X_test)
```

```python
_,axe = plt.subplots()
axe.plot(X_test[0])
axe.scatter([11], model.predict(X_test)[0], c='r')
axe.scatter([11], y_test)
```

```python
_,axe = plt.subplots()
axe.plot(hist.history['loss'])
```

```python
model = keras.Sequential([
#     layers.LSTM(11, input_shape=(None, 1)),
    layers.GRU(11, input_shape=(None, 1), return_sequences=True),
    layers.GRU(5),
    layers.Dense(1)
])

model.compile(loss='mse', optimizer=keras.optimizers.Adam(learning_rate=0.01))
hist = model.fit(X_train, y_train, epochs=100, verbose=0)
model.predict(X_test)
print('train', model.evaluate(X_train, y_train))
print('test', model.evaluate(X_test, y_test))
```

```python
#callback
mycb = [
#     keras.callbacks.LambdaCallback(on_epoch_begin=lambda epoch, logs: print('에포크 시작'))
    keras.callbacks.EarlyStopping(monitor='loss', min_delta=0.5, patience=3)
    
]

model = keras.Sequential([
    layers.SimpleRNN(11, input_shape=(None, 1)),
    layers.Dense(1)
])

model.compile(loss='mse', optimizer=keras.optimizers.Adam(learning_rate=0.01))
hist = model.fit(X_train, y_train, epochs=100, callbacks=mycb)

_,axe = plt.subplots()
axe.plot(hist.history['loss'])
```
