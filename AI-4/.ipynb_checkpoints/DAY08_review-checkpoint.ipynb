```python
from tensorflow import keras
from tensorflow.keras import layers

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.boston_housing.load_data()
# print(X_train.shape)
# print(y_train.shape)
# print(X_train[:5])
print(y_train[:5])
```

```python
print(len(y_train), len(y_test))
x = np.concatenate([y_train, y_test])
print(len(x))

s = pd.Series(x)
print(s.describe())

x_np = pd.cut(x, 11, labels=range(11)).to_numpy()
print(x_np[:5])
y_train_cat, y_test_cat = x_np[:404], x_np[404:]
print(len(y_train_cat), len(y_test_cat))
```

```python
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(13, ),
                 kernel_regularizer= keras.regularizers.L2(0.01)),   # keras.regularizers.L2 : 과적합 방지 작업
    layers.Dropout(0.1),     #  Dropout : 과적합 방지 ---->  10퍼센트 단위로 끊음?
    layers.Dense(64, activation='relu',
                kernel_regularizer= keras.regularizers.L2(0.01)),
    layers.Dropout(0.2),
    layers.Dense(1)
])

model.compile( #loss='mse',
            loss = keras.losses.MeanSquaredError(),
#             optimizer= 'adam',
            optimizer= keras.optimizers.Adam(learning_rate=0.01),
#             metrics=['accuracy']
            metrics=[keras.metrics.RootMeanSquaredError()])
#             metrics=[keras.metrics.RootMeanSquaredError(), keras.metrics.CosineSimilarity()])
hist = model.fit(X_train, y_train, epochs=50)  # verbose = 0  ===>  에포크 안보임

print('train evaluate', model.evaluate(X_train, y_train))
print('test evaluate', model.evaluate(X_test, y_test))
```

```python
# print(hist)
# print(hist.history)
for n in hist.history:
    print(n)
    
_, axe = plt.subplots()
axe.plot(hist.history['root_mean_squared_error'])
```

```python
#분류모델

y_train_cat2 = keras.utils.to_categorical(y_train_cat, 11)
y_test_cat2 = keras.utils.to_categorical(y_test_cat, 11)    # one-hot encoding
# print(y_test_cat[:5])
# print(y_test_cat2[:5])

model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(13, )),     #  Dropout : 과적합 방지 ---->  10퍼센트 단위로 끊음?
    layers.Dense(64, activation='relu'),
    layers.Dense(11, activation='softmax')   # softmax : 
#     layers.Dense(11)
])

model.compile(
#             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
#              loss=keras.losses.SparseCategoricalCrossentropy(),  # softmax가 되어 있으면 logits X 
             loss=keras.losses.CategoricalCrossentropy(),  # target, one-hot encoded
             optimizer=keras.optimizers.Adam(learning_rate=0.01),
             metrics=['accuracy'])
model.fit(X_train, y_train_cat2, epochs=50)


print('train evaluate', model.evaluate(X_train, y_train_cat2))
print('test evaluate', model.evaluate(X_test, y_test_cat2))
```
