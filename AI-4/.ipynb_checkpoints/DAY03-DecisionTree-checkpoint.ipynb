```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import precision_score

import pandas as pd

pd_data = load_breast_cancer(as_frame=True)
X = pd_data.data[['mean radius', 'mean texture', 'mean area']].values # to ndarray
y = pd_data.target.values

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=10)
```

```python
m = DecisionTreeClassifier(criterion='entropy').fit(X_train, y_train)
# print(y_test)
# print(m.predict(X_test))
# print(m.score(X_test, y_test))

plot_tree(m)
print(m.get_n_leaves())
print(m.get_depth())
```
