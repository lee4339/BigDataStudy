```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import completeness_score, homogeneity_score

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
X, y = load_iris(return_X_y=True)
# print(X)
sns.relplot(x=np.arange(len(y)), y=X[:,3], hue=y)
```

```python
m = KMeans(n_clusters=3).fit(X[:,3][:,np.newaxis])
```

```python
m.cluster_centers_
m.n_iter_
m.labels_
```

```python
sns.relplot(x=np.arange(len(y)), y=X[:,3], hue=m.labels_)
```

```python
print('completeness ', completeness_score(y, m.labels_))
print('homogeneity ', homogeneity_score(y, m.labels_))
```
