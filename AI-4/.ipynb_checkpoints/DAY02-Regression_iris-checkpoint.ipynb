```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# linear regression
# select class = 1
# petal width     vs.      sepal width
# plot regression line
```

```python
X, y = load_iris(return_X_y=True) # 1, 3

class_filter = y==1 # select class = 1
X = X[class_filter]

indep_v = X[:, 1][:, np.newaxis]
dep_v = X[:,3]

# print(indep_v)
# print(dep_v)

indep_train, indep_test, dep_train, dep_test = train_test_split(indep_v, dep_v)
```

```python
_, axe = plt.subplots()
axe.scatter(indep_train, dep_train)
```

```python
m = LinearRegression().fit(indep_train, dep_train)
print(m.coef_, m.intercept_)

_, axe = plt.subplots()
axe.scatter(indep_train, dep_train)
xs = np.linspace(indep_train.min(), indep_train.max(), 50)
axe.plot(xs, xs*m.coef_+m.intercept_,c='r')

print(r2_score(dep_test, indep_test * m.coef_ + m.intercept_))
```
