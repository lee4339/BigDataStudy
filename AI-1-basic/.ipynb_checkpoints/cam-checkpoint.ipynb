```python
with open('전국무인교통단속카메라표준데이터.csv', 'r') as f:
    raw = f.readlines()
```

```python
#설치연도 별로 카메라를 몇대의 카메라를 설치하였는가
raw_m = raw[0:3]

r1 = []
for line in raw_m:
    r1.append(line[:-1].split(','))
#print(r1)

r2 = []
for i, x_i in enumerate(r1[0]):
    print(i, x_i)
```

```python
r3 = []
for raw_line in raw:
    x = raw_line.split(',')
    r3.append(x[17])
#print(r3)


r4 = []
for n in r3:
    try:
        r4.append(int(n))
    except:
        pass
print(r4)
        
```
