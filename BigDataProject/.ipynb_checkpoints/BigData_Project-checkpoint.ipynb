```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

restaurant_data = pd.read_csv('Restaurant_information.csv', encoding='cp949')
covid19_data = pd.read_csv('covid19_data.csv')
delivery_data = pd.read_csv('Delivery_service_data.csv', encoding='cp949')

# print(restaurant_data.columns)
```

```python
#<restaurant_data> 데이터 전처리#  ---> 주소 통합

df_addr = restaurant_data[['소재지전체주소', '도로명전체주소']]
a = df_addr.iloc[:, 0].isna()
b = df_addr.iloc[:, 1].isna()
# print(a.value_counts())
# print(b.value_counts())

def addr_fn(x):
    try:
        s = x.split(' ')[0]
        if s[-1] == '시' or s[-1] == '도':
            return s
        else: 
            return np.NaN
    except:
        return np.NaN
    
addr_a = df_addr['소재지전체주소'].apply(addr_fn)
addr_b = df_addr['도로명전체주소'].apply(addr_fn)


x = addr_a.isna()
addr_a[x] = addr_b[x]

# print(addr_a)
# print(addr_a.value_counts())
# print(addr_a.isna().value_counts())    # 16개의 결측치

#################################################

restaurant_status = restaurant_data['상세영업상태명']
restaurant_permission_date = restaurant_data['인허가일자']
restaurant_closing_date = restaurant_data['폐업일자']
restaurant_coordinates_x = restaurant_data['좌표정보(x)']
restaurant_coordinates_y = restaurant_data['좌표정보(y)']

##################
#데이터프레임제작#
restaurant_df = pd.DataFrame()
restaurant_df['상세영업상태명'] = restaurant_status
restaurant_df['인허가일자'] = restaurant_permission_date
restaurant_df['폐업일자'] = restaurant_closing_date         # NaN 존재
restaurant_df['소재지'] = addr_a            
restaurant_df['좌표정보(x)'] = restaurant_coordinates_x     #NaN 존재
restaurant_df['좌표정보(y)'] = restaurant_coordinates_y     #NaN 존재


##############################################################
restaurant_df


# print(restaurant_df['폐업일자'].isna().value_counts())      # NaN : 691637개 --> 폐업이 아닌 음식점
# print(restaurant_df['좌표정보(x)'].isna().value_counts())   # NaN : 108733
# print(restaurant_df['좌표정보(y)'].isna().value_counts())   # NaN : 108733
```
