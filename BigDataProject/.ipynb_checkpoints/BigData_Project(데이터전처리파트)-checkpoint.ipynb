```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

restaurant_data = pd.read_csv('Restaurant_information.csv', encoding='cp949')
covid19_data = pd.read_csv('covid19_data.csv')
delivery_data = pd.read_csv('Delivery_service_data.csv', encoding='cp949')

# print(restaurant_data.columns)
```

```python
# 주소통합
def addr_fn(x):
    try:
        s = x.split(' ')[0]
        if s[-1] == '시' or s[-1] == '도':
            return s
        else: 
            return np.NaN
    except:
        return np.NaN

# 폐업일자, 인허가일자 데이터 정리#
def restaurant_fn2(x):
    if type(x) != type('abc') or len(x) != 8 :
        return np.NaN
    date_list = list(x)
    y = x[:4]
    m_i = int(x[4:6])
    d_i = int(x[6:])
    if m_i == 2:
        date_list[6:] = ['2','8']
    if m_i < 1 or m_i > 12:
        date_list[4:6] = ['0','1']
    if d_i < 1 or d_i > 30:
        date_list[6:] = ['0','1']
    return ''.join(date_list)


def restaurant_fn1(a):
    x = a.astype('str')
    date_list = []
    for i in x:
        if len(i) == 7:
            date_list.append("0")
        elif len(i) == 6:
            date_list.append("0") 
        elif len(i) == 4:
            date_list.append("0")
        elif len(i) == 8:
            date_list.append(restaurant_fn2(i))
        else:
            date_list.append(i)
    return date_list
        
```

```python
#<restaurant_data> 데이터 전처리#  ---> 주소 통합

df_addr = restaurant_data[['소재지전체주소', '도로명전체주소']]
a = df_addr.iloc[:, 0].isna()
b = df_addr.iloc[:, 1].isna()
# print(a.value_counts())
# print(b.value_counts())
addr_a = df_addr['소재지전체주소'].apply(addr_fn)
addr_b = df_addr['도로명전체주소'].apply(addr_fn)

x = addr_a.isna()
addr_a[x] = addr_b[x]
restaurant_address_data = addr_a

# print(addr_a)
# print(addr_a.value_counts())
# print(addr_a.isna().value_counts())    # 16개의 결측치

##############################################################################################
#restaurant_data 필요한 값#

restaurant_status_data = restaurant_data['상세영업상태명']
restaurant_permission_data = restaurant_data['인허가일자']
restaurant_closing_data = restaurant_data['폐업일자']
restaurant_coordinates_x = restaurant_data['좌표정보(x)']
restaurant_coordinates_x.replace(449161.628896747, 193959.150482967, inplace=True)
restaurant_coordinates_y = restaurant_data['좌표정보(y)']
restaurant_coordinates_y.replace('한식', 449161.628896747, inplace=True)
restaurant_coordinates_y = restaurant_coordinates_y.astype('float64')
##############################################################################################
# 폐업일자 데이터 정리#

# x = restaurant_closing_date.astype('str')
# print(x)
# x1 = x.apply(lambda a: len(a))
# print(x1.unique())
# print(x1.value_counts())

restaurant_closing_data.replace(np.NaN, 0, inplace=True)
restaurant_closing_data = restaurant_closing_data.astype('int64')

restaurant_series = pd.Series(restaurant_fn1(restaurant_closing_data))
restaurant_series.replace("0", np.NaN, inplace=True)

restaurant_closing_data = restaurant_series.astype('datetime64')
##############################################################################################
# 인허가일자 데이터 정리#

# x = restaurant_permission_date.astype('str')
# print(x)
# x1 = x.apply(lambda a: len(a))
# print(x1.unique())
# print(x1.value_counts())

restaurant_permission_data.replace(np.NaN, 0, inplace=True)

restaurant_series2 = pd.Series(restaurant_fn1(restaurant_permission_data))
restaurant_series2.replace("0", np.NaN, inplace=True)

restaurant_permission_data = restaurant_series2.astype('datetime64')
################################################################################################
#데이터프레임제작#
restaurant_df = pd.DataFrame()
restaurant_df['상세영업상태명'] = restaurant_status_data
restaurant_df['인허가일자'] = restaurant_permission_data
restaurant_df['폐업일자'] = restaurant_closing_data         # NaN 존재
restaurant_df['소재지'] = restaurant_address_data            
restaurant_df['좌표정보(x)'] = restaurant_coordinates_x     #NaN 존재
restaurant_df['좌표정보(y)'] = restaurant_coordinates_y     #NaN 존재


###############################################################################################
restaurant_df

# print(restaurant_df['폐업일자'].isna().value_counts())      # NaT : 691658 폐업이 아닌 음식점
# print(restaurant_df['좌표정보(x)'].isna().value_counts())   # NaN : 108733
# print(restaurant_df['좌표정보(y)'].isna().value_counts())   # NaN : 108733
```
