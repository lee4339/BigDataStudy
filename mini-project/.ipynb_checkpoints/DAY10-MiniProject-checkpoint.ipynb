목표: 아래와 같이 전처리된 데이터 확보. pickle 형식으로 저장하여 차후 사용이 용이하도록 함. 
* 전국 각 지역마다 2019년도 부터 2022 초 까지의 데이터를 조사하여 전기차에 대한 증가치 
* 현재 전기차 충전소가 각 지역마다 얼마나 분포되어있는가



```python
import pickle

#전기자동차 현황 데이터 전처리 부분
with open('한국전력공사_지역별 전기차 현황정보_20220429.csv', 'r', encoding='utf-8') as f:
    cur_raw = f.readlines()

# 데이터 정리
raw_list = []
for line in cur_raw:
    raw_list.append(line[:-1].split(','))

#지역 데이터 축출
area_list = raw_list[0][1:-1]

#날짜 데이터 축출
days_list = [x[0] for x in raw_list[1:]]

#현황 데이터 축출
data_list = [x[1:-1] for x in raw_list[1:]]

for int_data_list in data_list:
    for index, i in enumerate(int_data_list):
        int_data_list[index] = int(i)
        
print(area_list)        
print(days_list)        
print(data_list)

#전국 전기자동차 충전소위치 데이터 전처리 부분


#데이터 피클 저장
pickle_data = dict()
pickle_data['area'] = area_list
pickle_data['days'] = days_list
pickle_data['data'] = data_list
pickle_data['comment'] = '연도가 증가함에 전기자동차의 댓수가 증가함을 볼 수 있고 각 지역의 충전소를 비교하여 부족한 지역에 대해 조사'

# with open('전기자동차 증가와 충전소 현황.pickle', 'wb') as f:
#     pickle.dump(pickle_data, f)
```

```python
# 데이터파일 list로 정리
pass1 = []

for i in cur_raw[0:]:
    pass1.append(i[:-1].split(","))
    

# 데이터파일 list로 정리한거 필요값만 남기기
pass2 = []

for i in pass1[0:]:
    pass2.append(i[1:-1])
    

# list값에 숫자 문자열에서 int형으로 변경 후 2차원 리스트로 다시 쪼개기
pass3_mini = []

for n in pass2[1:]:
    for m in n[0:]:
        pass3_mini.append(int(m))
        
pass3 = []
for i in range(0, 646, 17):
    pass3.append(pass3_mini[i: i + 17])
    
    
    
# 각 지역별로 값 모아주기 // 각 인데스의 [0],[1],[2]....번 값 각기 모아서 재배치

pass4 = [[0 for _ in range(len(pass3))] for _ in range(len(pass3[0]))]

for index_i, counts in enumerate(pass3):
    for index_j, value in enumerate(counts):
        pass4[index_j][index_i] = value
        
        

        
# 각 지역별 2019년도 1월부터 2022년도 3월까지의 전기차 현황정보 
# 2022년도 부터 나오는 값을 리버스로 순서를 바꿔줌 

reverse_pass4 = []

print(pass2[0])

for i in pass4[0:]:
    i.reverse()
    print(i)
```

```python
adr_raw = response.content.decode('utf-8')
#print(adr_raw)

addr1, addr2 = 0, 0
addr_list = []
while True:
        addr1 = adr_raw.find('<addr>', addr2)
        if addr1 == -1:
            break
            
        addr1 += len('<addr>')
        addr2 = adr_raw.find('</addr>', addr1)
        addr_list.append(adr_raw[addr1:addr2])
    

#print(addr_list)
```

```python
addr_list2 = []
addr_list3 = []

for i in addr_list:
    addr_list2.append(i.split(' '))

#print(addr_list2)

for j in addr_list2:
    addr_list3.append(j[0])
    
print(addr_list3)
```
