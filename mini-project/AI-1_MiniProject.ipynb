목표: 아래와 같이 전처리된 데이터 확보. pickle 형식으로 저장하여 차후 사용이 용이하도록 함. 
* 전국 각 지역마다 2019년도 부터 2022 초 까지의 데이터를 조사하여 전기차에 대한 증가치 
* 현재 전기차 충전소가 각 지역마다 얼마나 분포되어있는가



```python
import pickle

#전기자동차 현황 데이터 전처리 부분
with open('한국전력공사_지역별 전기차 현황정보_20220429.csv', 'r', encoding='utf-8') as f:
    cur_raw = f.readlines()

# 데이터 정리
raw_list = []
for line in cur_raw:
    raw_list.append(line[:-1].split(','))

#지역 데이터 축출
area_list = raw_list[0][1:-1]

#날짜 데이터 축출
days_list = [x[0] for x in raw_list[1:]]

#현황 데이터 축출
data_list = [x[1:-1] for x in raw_list[1:]]

for int_data_list in data_list:
    for index, i in enumerate(int_data_list):
        int_data_list[index] = int(i)

# area_list.reverse
# print(area_list)        
# print(days_list)        
# print(data_list)
```

```python
#전국 전기자동차 충전소위치 데이터 전처리 부분
# 자료불러오기
import requests

url = 'http://apis.data.go.kr/B552584/EvCharger/getChargerInfo'
s_key = 'Oc+0oEXikbOn8irQ007JmQedWhaRAR+Z1ZMGmyDUeBwbvv8GIJgssZL0tVHpSwkmI2uuPRNz4vTIzPgFeBj3/Q=='
params ={'serviceKey' : s_key, 'pageNo' : '1', 'numOfRows' : '9999'}

response = requests.get(url, params=params)

# 데이터 주소 정리
adr_raw = response.content.decode('utf-8')

addr1, addr2 = 0, 0
addr_list = []
while True:
        addr1 = adr_raw.find('<addr>', addr2)
        if addr1 == -1:
            break
            
        addr1 += len('<addr>')
        addr2 = adr_raw.find('</addr>', addr1)
        addr_list.append(adr_raw[addr1:addr2])

# 데이터 주소 지역만 자르기
addr_list2 = []
addr_list3 = []

for i in addr_list:
    addr_list2.append(i.split(' '))

for j in addr_list2:
    addr_list3.append(j[0])

# 지역 set으로 중복된 값 처리 및 필요없는 값 없애기
set_list3 = list(set(addr_list3)) 

one_list = []
for item in set_list3:
    if item == '' or item == '97-32' or item == '제일오투마트' or item == '5층' or item == 'null' or item == '지상' or item == ',':
        pass
    else:
        one_list.append(item)

# 지역별 충전소 카운터
all_area_count_list = []

for location in one_list:
    n = addr_list3.count(location)
    all_area_count_list.append([location, n])
    
# print(all_area_count_list)
```

```python
#데이터 피클 저장
pickle_data = dict()
pickle_data['area'] = area_list
pickle_data['days'] = days_list
pickle_data['data'] = data_list
pickle_data['all_area_data'] = all_area_count_list
pickle_data['comment'] = '연도가 증가함에 전기자동차의 댓수가 증가함을 볼 수 있고 각 지역의 충전소를 비교하여 부족한 지역에 대해 조사'

with open('전기자동차 증가와 충전소 현황.pickle', 'wb') as f:
    pickle.dump(pickle_data, f)
```
